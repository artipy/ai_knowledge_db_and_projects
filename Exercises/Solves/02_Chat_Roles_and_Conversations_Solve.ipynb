```python
## –ß–∞—Å—Ç—å 4
### –ó–∞–¥–∞—á–∞ 4.1
sys_prompt = """–¢—ã –ø—Ä–µ–ø–æ–¥–∞–≤–∞—Ç–µ–ª—å –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è –Ω–∞ —è–∑—ã–∫–µ Python.

–¢–≤–æ–∏ –æ–±—è–∑–∞–Ω–Ω–æ—Å—Ç–∏:
- –û—Ç–≤–µ—á–∞—Ç—å –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã –∫—Ä–∞—Ç–∫–æ –∏ –ø–æ–Ω—è—Ç–Ω–æ
- –ü—Ä–∏–≤–æ–¥–∏—Ç—å –ø—Ä–∏–º–µ—Ä—ã –∫–æ–¥–∞
- –û–±—ä—è—Å–Ω—è—Ç—å —Å–ª–æ–∂–Ω—ã–µ –∫–æ–Ω—Ü–µ–ø—Ü–∏–∏ –ø—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏
- –ü–æ–º–æ–≥–∞—Ç—å —Å –æ—Ç–ª–∞–¥–∫–æ–π –∫–æ–¥–∞

–ù–∞ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –≤–æ–ø—Ä–æ—Å—ã –æ—Ç–≤–µ—á–∞–π –∫—Ä–∞—Ç–∫–æ –≤ –æ–¥–Ω–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ, –∞ –Ω–∞ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –≤–æ–ø—Ä–æ—Å—ã –ø—Ä–∏–≤–æ–¥–∏ –∫—Ä–∞—Ç–∫–∏–µ –ø—Ä–∏–º–µ—Ä—ã –∫–æ–¥–∞."""

questions = [
    '–ß—Ç–æ —Ç–∞–∫–æ–µ —Å–ø–∏—Å–æ–∫ –≤ Python?',
    '–ö–∞–∫ –¥–æ–±–∞–≤–ª—è—Ç—å –≤ –Ω–µ–≥–æ —ç–ª–µ–º–µ–Ω—Ç—ã?',
    "–ê –∫–∞–∫ —É–¥–∞–ª—è—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã –∏–∑ –Ω–µ–≥–æ?",
    "–ü–æ–∫–∞–∂–∏ –ø—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤—Å–µ—Ö —ç—Ç–∏—Ö –æ–ø–µ—Ä–∞—Ü–∏–π",
    "–°–ø–∞—Å–∏–±–æ! –¢–µ–ø–µ—Ä—å –æ–±—ä—è—Å–Ω–∏ —á—Ç–æ —Ç–∞–∫–æ–µ —Å–ª–æ–≤–∞—Ä—å"
]

messages = [{"role": "system", "content": sys_prompt}]
input_tokens_list = []
output_tokens_list = []

for question in questions:

    user_dict = {"role": "user", "content": question}
    messages.append(user_dict)

    response = client.chat.completions.create(
        model=os.getenv("BASE_MODEL_FLASH_LITE"),
        messages=messages
    )

    assistant_dict = {"role": "assistant", "content": response.choices[0].message.content}
    messages.append(assistant_dict)

    print(f"–í–æ–ø—Ä–æ—Å: {question}\n–û—Ç–≤–µ—Ç: {response.choices[0].message.content}\n")

    input_tokens_list.append(response.usage.prompt_tokens)
    output_tokens_list.append(response.usage.completion_tokens)

print(f'–í—Å–µ–≥–æ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ –∏—Å—Ç–æ—Ä–∏–∏: {len(messages)}')
print(f"–í—Ö–æ–¥–Ω—ã–µ —Ç–æ–∫–µ–Ω—ã: {response.usage.prompt_tokens}\n–í—ã—Ö–æ–¥–Ω—ã–µ —Ç–æ–∫–µ–Ω—ã: {response.usage.completion_tokens}\n–í—Å–µ–≥–æ —Ç–æ–∫–µ–Ω–æ–≤: {response.usage.total_tokens}")
```

```python
### –ó–∞–¥–∞—á–∞ 4.2
import pandas as pd

df = pd.DataFrame(
    columns=["–í—Ö–æ–¥–Ω—ã–µ —Ç–æ–∫–µ–Ω—ã", "–í—ã—Ö–æ–¥–Ω—ã–µ —Ç–æ–∫–µ–Ω—ã", "–°—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–ø—Ä–æ—Å–∞"],
    data=list(zip(input_tokens_list, output_tokens_list, [(inp * (0.15 / 1000000) + out * (0.6 / 1000000)) for inp, out in zip(input_tokens_list, output_tokens_list)]))
)

df
```

```python
## –ß–∞—Å—Ç—å 5
### –ó–∞–¥–∞—á–∞ 5.1
def limit_history(messages: list, max_pairs: int = None) -> list:
    """
    –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ—Ç –∏—Å—Ç–æ—Ä–∏—é –¥–∏–∞–ª–æ–≥–∞.

    Args:
        messages: –ø–æ–ª–Ω—ã–π —Å–ø–∏—Å–æ–∫ —Å–æ–æ–±—â–µ–Ω–∏–π
        max_pairs: –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä user-assistant –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è

    Returns:
        –£—Å–µ—á–µ–Ω–Ω—ã–π —Å–ø–∏—Å–æ–∫: [system] + –ø–æ—Å–ª–µ–¥–Ω–∏–µ max_pairs –ø–∞—Ä
    """

    if max_pairs is None:
        return messages
    system_message = messages[0]
    user_assistant_pairs = messages[1:]
    if max_pairs * 2 > len(user_assistant_pairs):
        return messages
    limited_pairs = user_assistant_pairs[-max_pairs * 2:]
    return [system_message] + limited_pairs

strategies = {
    "–ü–æ–ª–Ω–∞—è –∏—Å—Ç–æ—Ä–∏—è": None,  # –ë–µ–∑ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π
    "–ü–æ—Å–ª–µ–¥–Ω–∏–µ 5 –ø–∞—Ä": 5,
    "–ü–æ—Å–ª–µ–¥–Ω–∏–µ 3 –ø–∞—Ä—ã": 3
}

questions = [
    "–ß—Ç–æ —Ç–∞–∫–æ–µ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è?",
    "–ö–∞–∫–∏–µ —Ç–∏–ø—ã –¥–∞–Ω–Ω—ã—Ö –µ—Å—Ç—å –≤ Python?",
    "–ß—Ç–æ —Ç–∞–∫–æ–µ —É—Å–ª–æ–≤–Ω—ã–π –æ–ø–µ—Ä–∞—Ç–æ—Ä if?",
    "–ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç —Ü–∏–∫–ª for?",
    "–ß—Ç–æ —Ç–∞–∫–æ–µ —Ñ—É–Ω–∫—Ü–∏—è?",
    "–ö–∞–∫ –ø–µ—Ä–µ–¥–∞–≤–∞—Ç—å –∞—Ä–≥—É–º–µ–Ω—Ç—ã –≤ —Ñ—É–Ω–∫—Ü–∏—é?",
    "–ß—Ç–æ —Ç–∞–∫–æ–µ —Å–ø–∏—Å–æ–∫?",
    "–ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç —Å–ª–æ–≤–∞—Ä—å?",
    "–ß—Ç–æ —Ç–∞–∫–æ–µ –∫–ª–∞—Å—Å—ã?",
    "–û–±—ä—è—Å–Ω–∏ –Ω–∞—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ"
]

for strategy_name, max_pairs in strategies.items():
    print(f"–°—Ç—Ä–∞—Ç–µ–≥–∏—è: {strategy_name}\n")
    messages = [{"role": "system", "content": sys_prompt}]
    for question in questions:
        messages.append({"role": "user", "content": question})

        response = client.chat.completions.create(
            model=os.getenv("BASE_MODEL_PRO"),
            messages=messages
        )

        messages.append({"role": "assistant", "content": response.choices[0].message.content})

        # –û–≥—Ä–∞–Ω–∏—á–∏–≤–∞–µ–º –∏—Å—Ç–æ—Ä–∏—é –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å–æ —Å—Ç—Ä–∞—Ç–µ–≥–∏–µ–π
        messages = limit_history(messages, max_pairs=max_pairs)

        # print(f"–í–æ–ø—Ä–æ—Å: {question}\n–û—Ç–≤–µ—Ç: {response.choices[0].message.content}\n")
    print(f'–í—Å–µ–≥–æ —Å–æ–æ–±—â–µ–Ω–∏–π –≤ –∏—Å—Ç–æ—Ä–∏–∏ –ø–æ—Å–ª–µ —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏ "{strategy_name}": {len(messages)}\n')
    print(f'Input tokens: {response.usage.prompt_tokens}, Output tokens: {response.usage.completion_tokens}, Total tokens: {response.usage.total_tokens}\n')
    print('Total cost: $' + str((response.usage.prompt_tokens * (0.15 / 1000000) + response.usage.completion_tokens * (0.6 / 1000000))))
```
